```python
import requests
import bs4
import pandas as pd
 
url='https://geo.craigslist.org/iso/us'
 
res = requests.get(url)
res.raise_for_status()
soup = bs4.BeautifulSoup(res.text, 'html.parser')
 
city_sites = dict()
for a in soup.find_all("a")[1:-7]:
    city = a.string
    if city not in city_sites:
        city_sites[city] = a.get('href')[8:-15]
        
cities = pd.DataFrame(list(city_sites.items()), columns = ['CITY', 'CRAIGSLIST_SITE'])
cities
```
