```python
import requests
import bs4
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np 
url='https://geo.craigslist.org/iso/us'
 
res = requests.get(url)
res.raise_for_status()
soup = bs4.BeautifulSoup(res.text, 'html.parser')
 
# Collects all the cities in the US, with their associated CL site.
city_sites = dict()
for a in soup.find_all("a")[1:-7]:
    city = a.string
    if city not in city_sites:
        city_sites[city] = a.get('href')[8:-15]
        
cities = pd.DataFrame(list(city_sites.items()), columns = ['CITY', 'CRAIGSLIST_SITE'])

# Finds all the areas within that city.
def get_all_areas(site):
    response = requests.get(SITE_URL % site)
    response.raise_for_status()  # Something failed?
    soup = BeautifulSoup(response.content, 'html.parser')
    raw = soup.select('ul.sublinks li a')
    sites = {a.attrs['href'].rsplit('/')[1]: a.attrs['title'] for a in raw}
    return sites

SITE_URL = 'http://%s.craigslist.org'

## Actually carries out the scraping of sub-areas, city by city.
## I comment it out because it takes a long time to execute.
# areas = []
# for site in cities['CRAIGSLIST_SITE']:
#     area_list = get_all_areas(site)
#     if len(area_list) > 0:
#         areas.append(area_list)
#     else:
#         areas.append(np.nan)
#     print('{} > {}'.format(site, area_list))

cities['areas'] = areas
cities.to_csv('cities_and_areas.csv')
```
